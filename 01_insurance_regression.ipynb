```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
import urllib
urllib.request.urlretrieve('https://github.com/zbz95/Intro_to_ML/raw/main/insurance.csv', 'insurance.csv')
```

# Упражнение 1
Откройте файл insurance.csv

Поделите файл на три блока train, validation, test. Размер train-80% всех данных, validation 10% всех данных, test 10% всех данных

Стандартизируйте features во всех трех блоках используя StandardScaler

```python
df = pd.read_csv('insurance.csv')
```

```python
train,validation = train_test_split(df, test_size=0.2, shuffle=True)
validation, test = train_test_split(validation, test_size=0.5, shuffle=True)
```

```python
scaler = StandardScaler()
scaler.fit(train[['age','sex','bmi','children','smoker']])
```

```python
train[['age','sex','bmi','children','smoker']] = scaler.transform(train[['age','sex','bmi','children','smoker']])
validation[['age','sex','bmi','children','smoker']] = scaler.transform(validation[['age','sex','bmi','children','smoker']])
test[['age','sex','bmi','children','smoker']] = scaler.transform(test[['age','sex','bmi','children','smoker']])
```

# Упражнение 2

Постройте модель линейной регрессии(LinearRegression) на блоке train и посчитайте root_mean_squared_error на блоке validation

```python
model1=LinearRegression()
model1.fit(train[['age','sex','bmi','children','smoker']],train['expenses'])
```

```python
validation['prediction'] = model1.predict(validation[['age','sex','bmi','children','smoker']])
```

```python
rmse = root_mean_squared_error(validation['expenses'], validation['prediction'])
print(rmse)
```

# Упражнение 3
Постройте модели KNeighborsRegressor используя разные варианты числа соседей, k = [5,10,15,20,25,50]

Обучите все модели на train и посчитайте  root_mean_squared_error на блоке validation для каждого k

```python
for k in [5,10,15,20,25,50]:
  model = KNeighborsRegressor(n_neighbors=k)
  model.fit(train[['age','sex','bmi','children','smoker']],train['expenses'])

  validation[f'predictiion_{k}'] = model.predict(validation[['age','sex','bmi','children','smoker']])
  rmse = root_mean_squared_error(validation['expenses'], validation[f'predictiion_{k}'])

  print(f'Number of nearest neighbors = {k}. Root mean sqr error = {rmse}')
```
